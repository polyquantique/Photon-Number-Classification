```python
import os
#os.environ["OMP_NUM_THREADS"] = '1'

import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_formats=['svg']
#%load_ext autoreload
#%autoreload 2

from os import listdir
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from tqdm.notebook import tqdm

```

```python
def PCA_Kmean(X, max_cluster=40):

    pca = PCA(n_components=1)
    feature = pca.fit_transform(X)

    scores = []
    best_score = 0
    cluster_list = range(2,max_cluster)
    for cluster_number in tqdm(cluster_list , desc="Clusters") :
        clusters = KMeans(n_clusters=cluster_number, random_state=42, init="k-means++", n_init='auto').fit_predict(feature)
        scores.append(silhouette_score(feature, clusters))

        if best_score < scores[-1]:
            optimal_clusters = clusters
            best_score = scores[-1]

    plt.figure()
    plt.plot(cluster_list, scores)
    plt.xlabel("Number of cluster")
    plt.ylabel("Silhouette score")

    
    bins = np.linspace(min(feature), max(feature), 100).reshape(-1)
    unique_label = np.unique(optimal_clusters)
    out = []

    for label in unique_label:
        out.append(X[optimal_clusters == label])
        
    plt.figure()
    for index1, cluster in enumerate(out):
        plt.hist(cluster , bins, alpha = 0.5, label=f"{index1}")
```

```python
signal_size = 800
path = "Datasets/NIST 1 (short)/"
max_cluster = 20

X = np.fromfile(f"{path}{listdir(path)[-1]}",dtype=np.float16).reshape(-1,signal_size)
```

```python
PCA_Kmean(X)
```
